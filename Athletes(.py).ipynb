```python
df4.sort_values(by= ['height', 'weight'], ascending=[False, False]).head(100)
```

```python
for x in df4.index:
    if df4.loc[x, 'height'] == 0:
        df4.drop(x, inplace=True)
#loop through 'height' alongside the index(dropping rows looks at the axis(0)) and drop rows that has height = 0
```

```python
df4 = df4.reset_index(drop=True)
df4
```

```python
df4.info()
```

```python
sorted_items = df4.sort_values(by= 'height', ascending = False)
# In order to get the top 20 weighted people in a visualization first sort the items by weight from highest to lowest
```

```python
top_20 = sorted_items.head(20)
top_20
# then use the 'head' to get the top 20
```

```python
# then visualize it below
```

```python
top_20.plot(kind = 'barh', y = 'height', x = 'name', legend = False, color= 'green', title = 'Top 20 atletes with the most weight', xlabel = 'weight', figsize = (10,5))
```
